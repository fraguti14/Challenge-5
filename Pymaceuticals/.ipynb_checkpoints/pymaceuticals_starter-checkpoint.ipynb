### Analysis

**Key Observations:**

- Capomulin & Ramicane have the lowest mean on tumor volumes versul all other drug regimens.
    - This means thet are potentially effective in reducing tumor size.
- Capomulin & Ramicane have the smallest tumor volume variances & standard deviations. 
    - Hinting towars consistency and reliability in their results.
- Infubinol, Ketapril, Naftisol, Placebo, Propriva, Stelasyn, and Zoniferol have higher mean tumor volumes.
    - Meaning potentially less effectiveness in reducing tumor size versus Capomulin & Ramicane.
- The linear regression analysis shows a positive correlation between the average tumor volume and time. 
- The slope indicates that the tumor volume increases by 0.95 'mm^3' each day.
- The correlation coefficient of 0.84 indicates a strong positive relationship between time and tumor volume.
- The p-value is significantly small, meaning that the relationship between time and tumor volume is statistically significant.

**Other Notes:**
- Data was cleaned and a duplicate entry for mouse ID "g989" was removed from the analysed dataset.
- Total Mouse ID's analysed: 248
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_data = pd.merge(mouse_metadata, study_results, on='Mouse ID', how="outer")

# Display the data table for preview
merged_data.head(4)
```

```python
# Checking the number of mice.
unique_mice = merged_data['Mouse ID'].nunique()
print(f"Number of unique mice IDs: {unique_mice}")
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupe_mice = merged_data[merged_data.duplicated(subset=['Mouse ID', 'Timepoint'])]['Mouse ID'].unique()
print(dupe_mice)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_rows = merged_data[merged_data.duplicated(['Mouse ID', 'Timepoint'])]
print("Duplicate Rows:")
duplicate_rows.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

cleaned_data = merged_data[merged_data["Mouse ID"].isin(dupe_mice)==False]
unique_mice_count_cleaned = cleaned_data['Mouse ID'].nunique()
```

```python
# Checking the number of mice in the clean DataFrame.
# Checking the number of mice in the clean DataFrame.
print(f"Number of unique mice IDs after cleaning: {unique_mice_count_cleaned}")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

# grouping by drug regimen
drug_reg = cleaned_data.groupby('Drug Regimen')

# Math
mean_vol = drug_reg['Tumor Volume (mm3)'].mean()
median_vol = drug_reg['Tumor Volume (mm3)'].median()
var_vol = drug_reg['Tumor Volume (mm3)'].var()
std_vol = drug_reg['Tumor Volume (mm3)'].std()
sem_volume = drug_reg['Tumor Volume (mm3)'].sem()

# Creating df
summary_stats = pd.DataFrame({
    'Mean Tumor Volume': mean_vol,
    'Median Tumor Volume': median_vol,
    'Tumor Volume Variance': var_vol,
    'Tumor Volume Std. Dev.': std_vol,
    'Tumor Volume Std. Err.': sem_volume
}).reset_index()

# rounding values
#summary_stats=round(summary_stats,2)

# Display df
print("Summary stats for tumor vol by 'Drug Regimen':")
summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

# agg method

summary_agg = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Renaming columns
summary_agg = summary_agg.rename(columns={
    'mean': 'Mean Tumor Volume',
    'median': 'Median Tumor Volume',
    'var': 'Tumor Volume Variance',
    'std': 'Tumor Volume Std. Dev.',
    'sem': 'Tumor Volume Std. Err.'
})

#Rounding values
#summary_agg=round(summary_agg,2)

# Display df
print("Summary stats for tumor vol by 'Drug Regimen':")
summary_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_tmpnt_df = cleaned_data.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_vol_df = pd.merge(max_tmpnt_df, cleaned_data, on=['Mouse ID', 'Timepoint'])
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_reg = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for regimen in treatment_reg:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volumes = final_tumor_vol_df[final_tumor_vol_df['Drug Regimen'] == regimen]['Tumor Volume (mm3)']
    
    # add subset 
    tumor_vol_data.append(tumor_volumes)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_volumes.quantile([0.25, 0.5, 0.75])
    lower_quartile = quartiles[0.25]
    upper_quartile = quartiles[0.75]
    iqr = upper_quartile - lower_quartile
    
    lower_bound = lower_quartile - 1.5 * iqr
    upper_bound = upper_quartile + 1.5 * iqr
    outliers = tumor_volumes[(tumor_volumes < lower_bound) | (tumor_volumes > upper_bound)]
    
    print(f"Potential outliers for {regimen}: {outliers.tolist()}")
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# Filteing & data
capomulin_data = cleaned_data[cleaned_data['Drug Regimen'] == 'Capomulin']
avg_tumor_vol = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_w = capomulin_data.groupby('Mouse ID')['Weight (g)'].first()

# scatter plot
plt.scatter(mouse_w, avg_tumor_vol, marker='o', color='royalblue')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# correlation coefficient
correlation_coefficient = st.pearsonr(mouse_w, avg_tumor_vol)[0]

# linear regression
slope, intercept, r_value, p_value, std_err = st.linregress(mouse_w, avg_tumor_vol)

# regression equation
regression = f"y = {slope:.2f}x + {intercept:.2f}"

# scatter plot
plt.scatter(mouse_w, avg_tumor_vol, marker='o', color='royalblue')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# plot regression line
plt.plot(mouse_w, slope * mouse_w + intercept, color='red', linestyle='--', label=regression)

# display
print(f"The correlation between mouse weight and the average tumor volume is {round(r_value,2)}")
plt.show()
```
